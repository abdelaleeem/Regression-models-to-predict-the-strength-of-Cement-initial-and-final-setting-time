## Let's Remove The Outliers with Three Standard Deviations Method

```python
def remove_outliers_std(df):
    df_no_outliers = df.copy()
    numerical_columns = df.select_dtypes(include='number').columns
    
    for col in numerical_columns:
        mean = df[col].mean()
        std_dev = df[col].std()
        lower_bound = mean - 3 * std_dev
        upper_bound = mean + 3 * std_dev
        
        # Filter the data
        mask = (df_no_outliers[col] >= lower_bound) & (df_no_outliers[col] <= upper_bound)
        df_no_outliers = df_no_outliers[mask]
    
    return df_no_outliers

df_no_outliers = remove_outliers_std(df)
```

## Let's Remove the features that's highly correlated with the other

```python
correlation_matrix = df_no_outliers.corr().abs()

# Choose the threshold for correlation
threshold = 0.7

# Find index pairs where the correlation is above the threshold
high_corr_var = np.where(correlation_matrix > threshold)
high_corr_var = [(correlation_matrix.columns[x], correlation_matrix.columns[y]) 
                 for x, y in zip(*high_corr_var) if x != y and x < y]
print(high_corr_var)
# Drop one of the features in each highly correlated pair
for x, y in high_corr_var:
    if x in df_no_outliers.columns:
        df_no_outliers.drop(x, axis=1, inplace=True)
```

## we will try L1 Regularization (LASSO Regression) and L2 Regularization (RIDGE Regression) to estimate the optimum Hyper Parameters

## Lasso work on feature selection beside overfitting problem as you may see that some of the features' coefficients may go down to 0 

```python
# List of alpha values for tuning
alphas = np.logspace(-4, 5, 50)

# Store results
ridge_train_errors = []
ridge_test_errors = []
lasso_train_errors = []
lasso_test_errors = []

# Trial and error approach for both Ridge and Lasso
for alpha in alphas:
    # Ridge
    ridge = Ridge(alpha=alpha, random_state=42)
    ridge.fit(x_train_scaled, y_train)
    ridge_train_errors.append(mean_absolute_error(y_train, ridge.predict(x_train_scaled)))
    ridge_test_errors.append(mean_absolute_error(y_test, ridge.predict(x_test_scaled)))
    
    # Lasso
    lasso = Lasso(alpha=alpha, random_state=42)
    lasso.fit(x_train_scaled, y_train)
    lasso_train_errors.append(mean_absolute_error(y_train, lasso.predict(x_train_scaled)))
    lasso_test_errors.append(mean_absolute_error(y_test, lasso.predict(x_test_scaled)))

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(alphas, ridge_train_errors, label="Ridge Train Error")
plt.plot(alphas, ridge_test_errors, label="Ridge Test Error")
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('Mean Absolute Error')
plt.title('Ridge Regression')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(alphas, lasso_train_errors, label="Lasso Train Error")
plt.plot(alphas, lasso_test_errors, label="Lasso Test Error")
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('Mean Absolute Error')
plt.title('Lasso Regression')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
min_idx_lasso = np.argmin(lasso_test_errors)
lasso_error_test = lasso_test_errors[min_idx_lasso]
lasso_error_train= lasso_train_errors[min_idx_lasso]

min_idx_ridge = np.argmin(ridge_test_errors)
ridge_error_test = ridge_test_errors[min_idx_ridge]
ridge_error_train= ridge_train_errors[min_idx_ridge]

print(f"lasso_error_test : {lasso_error_test} , lasso_error_train : {lasso_error_train} , optimum alpha : {alphas[min_idx_lasso]}")
print(f"ridge_error_test : {ridge_error_test} , ridge_error_train : {ridge_error_train} , optimum alpha : {alphas[min_idx_ridge]}")
```

Lasso Performs well with a MAE of 0.62 on test set

```python
best_model = Lasso(alpha = alphas[min_idx_lasso], random_state=42)
best_model.fit(x_train_scaled , y_train)
```

## I will try out if any polynomial degree will work "JUST FOR FUN"

```python
def plot_train_cv_mses(degrees, train_mses, cv_mses, title , degrees_ = range(1,11)):
    degrees = degrees_
    plt.plot(degrees, train_mses, marker='o', c='r', label='training MSEs'); 
    plt.plot(degrees, cv_mses, marker='o', c='b', label='CV MSEs'); 
    plt.title(title)
    plt.xlabel("degree"); 
    plt.ylabel("MAE"); 
    plt.legend()
    plt.show()
```

```python
# Initialize lists containing the lists, models, and scalers
train_maes = []
cv_maes = []
models = []
scalers = []

# Loop over 10 times. Each adding one more degree of polynomial higher than the last.
for degree in range(1,11):
    
    # Add polynomial features to the training set
    poly = PolynomialFeatures(degree, include_bias=False)
    X_train_mapped = poly.fit_transform(x_train)
    scaler  = StandardScaler()
    X_train_mapped = poly.fit_transform(x_train)
    X_train_mapped_scaled = scaler.fit_transform(X_train_mapped)

    # Create and train the model
    model = LinearRegression()
    model.fit( X_train_mapped_scaled, y_train)
    models.append(model)
    
    # Compute the training MSE
    yhat = model.predict(X_train_mapped_scaled)
    train_mae = mean_absolute_error(y_train, yhat)
    train_maes.append(train_mae)
    
    # Add polynomial features and scale the cross validation set
    poly = PolynomialFeatures(degree, include_bias=False)
    X_cv_mapped = poly.fit_transform(x_test)
    X_cv_mapped_scaled = scaler.fit_transform(X_cv_mapped)
    # Compute the cross validation MSE
    yhat = model.predict(X_cv_mapped_scaled)
    cv_mse = mean_absolute_error(y_test, yhat)
    cv_maes.append(cv_mse)
    
# Plot the results
degrees=range(1,6)
plot_train_cv_mses(degrees, train_maes, cv_maes, title="degree of polynomial vs. train and CV MSEs")
```

```python
error_train = []
error_test  = [] 
estimators = range(10 , 510 , 1)
for estimator in estimators:
    model = XGBRegressor(n_estimators = estimator ,random_state = 42)
    model.fit(x_train , y_train)

    y_pred_train = model.predict(x_train)
    y_pred_test  = model.predict(x_test)

    error_train.append(mean_absolute_error(y_train , y_pred_train))
    error_test.append(mean_absolute_error(y_test , y_pred_test))

plt.plot(estimators , error_train , color = "blue" , label = "train acc")
plt.plot(estimators , error_test , color = "red" , label = "test acc")
plt.xlabel("n_estimators")
plt.ylabel("MAE")
plt.legend()

print(f"train error {error_train[np.argmin(error_test)]}")
print(f"test error {error_test[np.argmin(error_test)]}")
print(f"best estimator is {estimators[np.argmin(error_test)]}")
```
